```python
#Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import requests
from config import api_key
```

```python
latitude = np.random.uniform(low=-90.000,high=90.000,size=2000)
longitude = np.random.uniform(low=-180.000,high=180.000,size =2000)
lat_lngs = zip(latitude,longitude)
lat_lngs
```

```python
coordinates =list(lat_lngs)
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID="+api_key
```

```python
from citipy import citipy
```

```python
city_list=[]
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    if city not in city_list:
        city_list.append(city)
len(city_list)
```

```python
city_url = url + "&q=" + city.replace(" ","+")
```

```python
city_weather = requests.get(city_url).json()
city_weather
```

```python
weather_data =[]
record_count=1
set_count=1
for i, city in enumerate(city_list):
    if (i%50==0 and i>=50):
        set_count+=1
        record_count = 1
    city_url = url + "&q=" + city.replace(" ","+")
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    record_count+=1
    try:
        weather = requests.get(city_url).json()
        lat=weather["coord"]["lat"]
        lng=weather["coord"]["lon"]
        max_temp=weather["main"]["temp_max"]
        humidity= weather["main"]["humidity"]
        clouds = weather["clouds"]["all"]
        wind = weather["wind"]["speed"]
        country=weather["sys"]["country"]
        date = datetime.utcfromtimestamp(weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        weather_data.append({"city":city.title(),
                           "lat":lat,
                           "lng":lng,
                            "max_temp":max_temp,
                            "humidity":humidity,
                            "cloudiness":clouds,
                            "wind speed":wind,
                            "country":country,
                            "date":date})
    except:
        print("ERROR")
        pass
    
print("-----------------------")
print("Data Retrieval Complete")
print("-----------------------")
print(len(weather_data))
    
```

```python
city_df = pd.DataFrame(weather_data)
new_order=["city","country","date","lat","lng","max_temp","humidity","cloudiness","wind speed"]
city_df = city_df[new_order]
city_df.head(4)
```

```python
output_file ="WeatherPy_database.csv"
city_df.to_csv(output_file,index_label="City_ID")
```
